```python
# Example usage
# input_list = ["hey", "how", "is", "hey", "you", "how", "are", "you", "hey"]
# search_elem = "you"

def search(input_list, search_elem):
    if search_elem in input_list:
        return input_list.index(search_elem)
    else:
        return -1  # Return -1 if the element is not found

# Call the Function
input_list = ["hey", "how", "is", "hey", "you", "how", "are", "you", "hey"]
search_elem = "you"
output = search(input_list, search_elem)
print(output)
```

```python
def count(input_list, search_elem):
    return input_list.count(search_elem)

# Call the Function
input_list = ["hey", "how", "is", "hey", "you", "how", "are", "you", "hey"]
search_elem = "hey"
output = count(input_list, search_elem)
print(output)
```

```python
def search_location(input_list, search_list):
    output_dict = {term: [] for term in search_list}  # Initialize the dict with empty lists

    for term in search_list:
        for i in range(len(input_list)):  # Iterate through indices
            if input_list[i] == term:  # Check if the current value matches the term
                output_dict[term].append(i)  # Append the index to the corresponding list

    return output_dict

# Call the Function
input_list = ["hey", "how", "is", "hey", "you", "how", "are", "you", "hey"]
search_list = ["hey", "you"]
output = search_location(input_list, search_list)
print(output)
```
